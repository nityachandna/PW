#### Q2) What is a scatter plot? Use the following code to generate data for x and y. 
#### Using this generated data, plot a scatter plot:
#### import numpy as np
#### np.random.seed(3)
#### x = 3 + np.random.normal(0, 2, 50)
#### y = 3 + np.random.normal(0, 2, len(x))
#### Note: Also add title, xlabel, and ylabel to the plot.

```python
A scatter plot is a visual representation of how two variables relate to each other. 
It can be used to explore the relationship between two variables.
It is a diagram where each value in the data set is represented by a dot.
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y,color="red")
plt.xlabel("x")
plt.ylabel("y")
plt.title("scatter plot")
```

```python
The subplot() function provides a way to plot multiple plots on a single figure.
Given the number of rows and columns, it returns a tuple (fig, ax), giving,
a single figure 'fig' with an array of axes 'ax'.
```

```python
import numpy as np
```

```python
x1 = np.array([0, 1, 2, 3, 4, 5]) 
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x1,y1,color="r")

x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2= np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x2,y2,color="green")

x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x3,y3,color="blue")

x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x4,y4,color="orange")
```

```python
Bar plots are a type of data visualization used to represent data in the form of rectangular bars.
The height of each bar represents the value of a data point.
The width of each bar represents the category of the data. 
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit,color="magenta")
```

#### Q5) What is a box plot? Why is it used? Using the following data plot a box plot.
#### box1 = np.random.normal(100, 10, 200)
#### box2 = np.random.normal(90, 20, 200)

```python
A Box Plot is also known as Whisker Plot, created to display the summary of the set of data values
having properties like minimum, first quartile, median, third quartile and maximum. 
In the box plot, a box is created from the first quartile to the third quartile,
a vertical line is also there which goes through the box at the median. 
Here x-axis denotes the data to be plotted while the y-axis shows the frequency distribution.
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
